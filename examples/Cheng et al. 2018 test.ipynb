```python
from ibis import IBIS_Main
```

```python
_META = dict(
        # depth handling
        depth_increases_down=True,
        depth_tol=0.0,  # 0 => no rounding

        # boutique sampler
        fraction_det=0.2,
        r02_clip_lo=0.001,
        r02_clip_hi=1000.0,

        # r02 max bound from data (IMPORTANT)
        r02_max_k_sigma=3.0,  # use (r08+kσ)/(r28-kσ)

        # r02 relative uncertainty proposal
        r02_err_mode="halfnorm",   # "fixed"|"uniform"|"halfnorm"
        r02_rel_err=0.25,
        r02_rel_min=0.005,
        r02_rel_max=1.0,

        # validity gates
        age_sigma_gate=3.0,
        age_eps=1e-12,

        # sigma cheating prevention (EB): default False
        use_r02_in_sigma=True,

        # batch sampler controls
        batch_size = 1000,
        max_batches=5000,
        keep_frac=0.08,
        keep_factor=8,

        # layer windows
        neighbors=1,

        # pair weighting
        pair_weight_mode="exp",   # "none"|"exp"|"power"
        pair_tau_layers=2.0,
        pair_alpha=1.0,
        min_pair_weight=1e-6,

        # prior mixture
        global_mass=0.7,
        n_samples_global=4000,
        n_samples_layer=2500,

        # KDE fit controls
        kde_beta=0.7,
        bw_grid=None,
        cv=5,
        bw_subsample=3000,
        grid_n=1200,
        hi_pct=99.9,
        floor_pdf=1e-12,
    Verbose = True
    )
```

```python
mm = IBIS_Main.IBIS('/Users/johncarter/Documents/IBIS/IBIS paleo/Cheng_2018_MSD_data_clean_up.xlsx',
    sample_name='Cheng2018_MSD_ibis',          
    MCMC_samples=2000000,
    MCMC_burn_in=1000000,
    MCMC_Strat_samples=3000000,
    n_chains=3,
    Start_from_pickles=True, 
    show_bird=False, 
    method='thoth',
    strat_resolution=100, 
                    diction_meta= _META
)
 
```

```python
# Initial thorium correction by Cheng et al. (2018)
"""
	U decay constants: l238 = 1.55125x10-10 (Jaffey et al., 1971) and l234 = 2.82206x10-6 (Cheng et al., 2013). Th decay constant:  l230 = 9.1705x10-6 (Cheng et al., 2013).						
	*d234U = ([234U/238U]activity – 1)x1000.  ** d234Uinitial was calculated based on 230Th age (T), i.e., d234Uinitial = d234Umeasured x el234xT.  						
	***B.P. stands for “Before Present” where the “Present” is defined as the year 1950 A.D.  						
	Corrected 230Th ages assume the initial 230Th/232Th atomic ratio of 2.5  ±1.5 x10-6.						
	MSD-1-M2-17 and MSD1-B-5 are not used due to large errors.						
							
"""
cheng2018_corr_atm = np.array([2.5*1e-6, 1.5*1e-6])

def convert_to_activty(r, r_err):

    lam_230 = np.log(2)/75000
    lam_232 = np.log(2)/1.4e10
    act_ratio = r * (lam_230/lam_232)

    act_ratio_err = np.sqrt(((lam_230/lam_232)**2 * r_err**2))

    return act_ratio, act_ratio_err

act_corr_cheng, act_corr_cheng_err = convert_to_activty(cheng2018_corr_atm[0], cheng2018_corr_atm[1])
```

# IBIS
IBIS correction utilizing a full inversion of the dataset is often more precise than the chosen correction by Cheng et al. (2018) for stalagmite MSD. Consequently, inferred ages are more precise-> and could lead to more precision in the calibration of the radiocarbon curve.
